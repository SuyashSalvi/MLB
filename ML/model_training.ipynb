```python
import pandas as pd

# Load the dataset
df = pd.read_csv("gs://mlb-prospect-data/cleaned_mlb_homeruns.csv")

# Ensure WAR is numeric
df["WAR"] = pd.to_numeric(df["WAR"], errors="coerce")

# Sort data for proper shifting
df = df.sort_values(by=["PlayerName", "Year"])

# Compute rolling WAR average (past 3 years)
df["WAR_3yr_Avg"] = df.groupby("PlayerName")["WAR"].rolling(window=3, min_periods=1).mean().reset_index(0, drop=True)

# Compute WAR improvement rate (difference from last year)
df["WAR_Change"] = df.groupby("PlayerName")["WAR"].diff()

# Drop NaN values (for first-year players)
df = df.dropna(subset=["WAR_3yr_Avg", "WAR_Change"])

# Save the updated dataset
df.to_csv("gs://mlb-prospect-data/enhanced_mlb_homeruns.csv", index=False)
print("âœ… Dataset updated with historical WAR trends!")
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Select key performance features
features = ["ExitVelocity", "HitDistance", "LaunchAngle", "WAR_3yr_Avg", "WAR_Change"]

# Normalize Data (to remove scale bias)
df_normalized = (df[features] - df[features].mean()) / df[features].std()

# Compute similarity scores
similarity_matrix = cosine_similarity(df_normalized)

# Function to get top similar players
def get_similar_players(player_name, top_n=3):
    idx = df[df["PlayerName"] == player_name].index[0]
    sim_scores = list(enumerate(similarity_matrix[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)[1:top_n+1]
    similar_players = [df.iloc[i[0]]["PlayerName"] for i in sim_scores]
    return similar_players

# Example Usage
print("Similar Players to Shohei Ohtani:", get_similar_players("Shohei Ohtani"))
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from google.cloud import storage

# Prepare Data for Time-Series Prediction
players = df["PlayerName"].unique()
sequence_length = 3  # Use past 3 years of data

X, y = [], []
for player in players:
    player_data = df[df["PlayerName"] == player][["WAR"]].values
    for i in range(len(player_data) - sequence_length):
        X.append(player_data[i:i+sequence_length])
        y.append(player_data[i+sequence_length])

X, y = np.array(X), np.array(y)

# Build LSTM Model
model = Sequential([
    LSTM(50, activation="relu", input_shape=(sequence_length, 1), return_sequences=True),
    LSTM(50, activation="relu"),
    Dense(1)
])

model.compile(optimizer="adam", loss="mse")

# Train Model
model.fit(X, y, epochs=50, batch_size=16)

# âœ… Save the model locally first
local_model_path = "war_forecast_lstm.h5"
model.save(local_model_path)

# âœ… Upload the model to Google Cloud Storage (GCS)
def upload_to_gcs(local_path, bucket_name, gcs_path):
    storage_client = storage.Client()
    bucket = storage_client.bucket(bucket_name)
    blob = bucket.blob(gcs_path)
    blob.upload_from_filename(local_path)
    print(f"âœ… Model uploaded to GCS: gs://{bucket_name}/{gcs_path}")

# Define GCS Bucket and Path
bucket_name = "mlb-prospect-data"
gcs_model_path = "war_forecast_lstm.h5"

# Upload
upload_to_gcs(local_model_path, bucket_name, gcs_model_path)
```

```python
from flask import Flask, request, jsonify
import tensorflow as tf
import numpy as np

app = Flask(__name__)

# Load the trained LSTM model
model = tf.keras.models.load_model("gs://mlb-prospect-data/war_forecast_lstm.h5")

@app.route("/predict", methods=["POST"])  # âœ… Ensure POST method is allowed
def predict():
    if request.method == "POST":
        try:
            data = request.get_json()  # âœ… Use get_json() instead of request.json
            if not data or "WAR_History" not in data:
                return jsonify({"error": "Missing 'WAR_History' in request"}), 400
            
            # Ensure input shape matches LSTM model expectations
            input_war = np.array(data["WAR_History"]).reshape(1, -1, 1)  
            predicted_war = model.predict(input_war)[0][0]

            return jsonify({"predicted_future_WAR": predicted_war})

        except Exception as e:
            return jsonify({"error": str(e)}), 500  # âœ… Return proper error messages

# Run Flask App
if __name__ == "__main__":
    app.run(host="0.0.0.0", port=5000, debug=True)  # âœ… Ensure the app runs on 0.0.0.0
```
